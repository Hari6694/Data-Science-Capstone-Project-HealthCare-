```python
# Here Non-Diabetics patients are more than Diabetic patients which could lead to model baisness to one output, 
# to overcome the issue we are using here SMOTE class to balanced the outcomes variable so that the model could get its accuracy without baisness.

from imblearn.over_sampling import SMOTE         
sm=SMOTE(random_state=42)
x_resample,y_resample=sm.fit_resample(x,y)
```

```python
# Diving the dataset into training and testing set for the predict by various model further.

from sklearn.model_selection import train_test_split
```

```python
x_train,x_test,y_train,y_test=train_test_split(x_resample,y_resample,test_size=0.2,random_state=0)
```

```python
print(x_train.shape)
print(x_test.shape)
print(y_train.shape)
print(y_test.shape)
```

###2. Apply an appropriate classification algorithm to build a model.


```python
#Evaluating metrics to check the model accuracy by various scores

from sklearn.metrics import confusion_matrix, classification_report, accuracy_score, precision_score,f1_score,recall_score,roc_curve,roc_auc_score
```

#Logistic Regression Model


```python
from sklearn.linear_model import LogisticRegression
```

```python
logestic_reg=LogisticRegression() 
```

```python
logestic_reg.fit(x_train,y_train)
```

```python
y_pred=logestic_reg.predict(x_test)
```

```python
# Confusion matrix

cm=confusion_matrix(y_test,y_pred)
print(cm)
```

```python
accuracy_score(y_test,y_pred)
```

```python
print(classification_report(y_test,y_pred))
```

###4. Create a classification report by analyzing sensitivity, specificity, AUC (ROC curve), etc.

```python
#logistic regression 
logestic_reg.fit(x_train,y_train)
# predict probabilities
y_pred_probs = logestic_reg.predict(x_train)
# keep probabilities for the positive outcome only
#probs = probs[:, 1]
# calculate AUC
auc = roc_auc_score(y_train, y_pred_probs)
print('AUC: %.3f' % auc)
# calculate roc curve
fpr, tpr, thresholds = roc_curve(y_train, y_pred_probs)
# plot no skill
plt.plot([0, 1], [0, 1], linestyle='--')
# plot the roc curve for the model
plt.plot(fpr, tpr, marker='.')
plt.show()
```
